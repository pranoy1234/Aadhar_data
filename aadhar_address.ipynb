```python
from PIL import Image
import pytesseract, re
import cv2
import numpy as np
from fuzzywuzzy import fuzz




#noise removal
#--------------------------
# Read image with opencv
img = cv2.imread("/home/user/project/git/Aadhar_data/data/aadhar_card/a6.png")

# Convert to gray
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply dilation and erosion to remove some noise
kernel = np.ones((1, 1), np.uint8)
img = cv2.dilate(img, kernel, iterations=1)
cv2.imwrite("./dilate/dilate.jpg", img)

img = cv2.erode(img, kernel, iterations=1)
cv2.imwrite("./erode/erode.jpg", img)
# Write image after removed noise
cv2.imwrite("removed_noise.jpg", img)

#  Apply threshold to get image with only black and white
img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,23, 70)

# Write the image after apply opencv to do some ...
cv2.imwrite("thres.jpg", img)


f = "thres.jpg"
config="--psm 01"

t = pytesseract.image_to_string(Image.open(f),lang='eng',config=config)

#print(t)

print('\n'*3)
start=-1
end=-1
add=''
t=t.split('\n')



a_file = open("states.txt")
lines = a_file.readlines()

#finding Address
#------------------------------------------------------------

for i in range(len(t)):
    
    #applying fuzzy logic to get the address keyword
    #----------------------------------------------------
    
    for word in (t[i].split(' ')):
        if fuzz.ratio(word.lower(),'address')>70 :
            start=i
            break

for i in range(start,len(t)):
    word_list=t[i].split(' ')
    for j in range(len(word_list)-1):
         
        if word_list[j] =='Aadhaar' or word_list[j] =='Adhikar' or word_list[j] =='Aadmi' or word_list[j] =='Aam' or word_list[j]=='Admi':
            continue
        else:
            for line in lines:
                w_l=line.split(' ')
                
                #if len(w_l)==1:
                if fuzz.ratio(word_list[j].lower(),line.lower())>60 :
                    end=i
                    break
                if len(w_l)>1:
                    for k in range(len(w_l)-1):
                        if fuzz.ratio(word_list[j].lower(),w_l[k].lower())>50 and fuzz.ratio(word_list[j+1].lower(),w_l[k+1].lower())>50:
                            end=i
                            #print('doubleeeeeeee',i,word_list[j],w_l[k],word_list[j+1],w_l[k+1])
                            
                            

#Finding Name:
#----------------------------------

end1=-1
end2=-1
for i in range (len(t)):
    for word in (t[i].split(' ')):
        
        if word.lower().find('yob') !=-1 or word.lower().find('dob') !=-1 or word.lower().find('y ob') != -1 or word.lower().find('d ob')!=-1 or word.lower().find('y06')!=-1 or word.lower().find('year')!=-1 or word.lower().find('oob')!=-1:
            end1=i
            
    for word in (t[i].split(' ')):
        
        if word.lower().find('father:') !=-1 or word.lower().find('father')!=-1:
            end2=i
            
print('fetched name= ')
print('---------------------------')
print()          

if end2!=-1:
    print(t[end2-2])
if end1 !=-1 and end2==-1:
    if end1==1:
        print(t[0])
    for i in range(end1-1,0,-1):
        if t[i] == '' or t[i]==' ':
            continue
        else:
            print(t[i])
            break



print()
print('fetched YOB= ')
print('---------------------------')
print()
if end1 != -1:
    print(t[end1])


print()                    
print('fetched address= ')
print('---------------------------')
print()
if start !=-1 and end !=-1 and start<end:
    for i in range(start,end+1):
        print(t[i])
if start==-1 and end !=-1:
    for i in range(end-8,end+1):
        print(t[i])
    
        
        
```
