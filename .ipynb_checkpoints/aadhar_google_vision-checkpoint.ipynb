```python
import json
from fuzzywuzzy import fuzz

#removing the term 'e-mail address'
#--------------------------------------

def remove_email_address(text):
    
    for i in range(len(text)):
        word_list=text[i].split(' ')
        flag=0
        for word in word_list:
            if word=='e-mail':
                word_list.remove('e-mail')
                update_sen=''
                for t in word_list:
                    update_sen=update_sen+ ' '+t
                text[i]=update_sen
                word_list=text[i].split(' ')
                #checking address term in the next word

                if word=='address':    
                    word_list.remove('address')
                    update_sen=''
                    for t in word_list:
                        update_sen=update_sen+ ' '+t
                        text[i]=update_sen
                    flag=1

                if flag==0:
                    #checkimg address keyword in the immediate next line
                    if i+1 <len(text)-2:
                        word_list_nested=text[i+1].split(' ')
                        for word1 in word_list_nested:
                            if word1=='address.' or word1=='address':

                                word_list_nested.remove(word1)
                                update_sen=''
                                for t in word_list_nested:
                                    update_sen=update_sen+ ' '+t
                                    text[i+1]=update_sen

    return text

def address_generator(text):
    
    #finding start Index
    #---------------------
    start=-1
    for i in range(0,len(text)):
        flag=0
        for word in text[i].split(' '):
            if len(word)>2:
                if (word[0]=='a' or word[0]=='A') and word[1]=='d':
                    if fuzz.ratio(word.lower(),'address:') >=55:
                        start=i
                        flag=1
                        break
        if flag==1:
            break
            
    #finding end Index
    #---------------------
    end=-1
    a_file = open("states.txt")
    lines = a_file.readlines()
    flag=0
    for i in range(start,len(text)):
        word_list=text[i].split(' ')
        #print(word_list)
        for j in range(len(word_list)-1):
            if word_list[j] =='Aadhaar' or word_list[j] =='Adhikar' or word_list[j] =='Aadmi' or word_list[j] =='Aam' or word_list[j]=='Admi':
                continue
            else:
                for line in lines:
                    w_l=line.split(' ')
                    if len(w_l)==1:
                        if fuzz.ratio(word_list[j].lower(),line.lower())>67 :
                            end=i
                            flag=1
                            break
                        
                        if j==len(word_list)-2:
                            if fuzz.ratio(word_list[j+1].lower(),line.lower())>67 :
                                end=i
                                flag=1
                        
                                
                        
                    if len(w_l)>1 :
                        for k in range(len(w_l)-1):
                            if fuzz.ratio(word_list[j].lower(),w_l[k].lower())>50 and fuzz.ratio(word_list[j+1].lower(),w_l[k+1].lower())>50:
                                end=i
                                flag=1
                                break
                            if j==len(word_list)-2:
                                if fuzz.ratio(word_list[j+1].lower(),w_l[k].lower())>80:
                                    
                                    end=i+1
                                    flag=1
                                    break
                                    
        if flag==1:
            break                    
    return text,start,end

def name_dob_generator(text):
    dob_idx=-1
    name_idx=-1
    
    #finding name wrt to dob
    #------------------------------
    for i in range(len(text)):
        flag=0
        word_list=text[i].split(' ')
        for word in word_list:
            if word.find('YOB')!=-1:
                dob_idx=i
                flag=1
                break
            if word.find('DOB')!=-1:
                dob_idx=i
                flag=1
                break
            if word.find('Year')!=-1 or word.find('Birth')!=-1:
                dob_idx=i
                flag=2
                break
                
        up_n_idx=False        
        if flag!=0:
            if flag==1:
                if dob_idx>4:
                    temp=text[dob_idx-1].split()
                    for w in temp:
                        if w.find('Father')!=-1 or w.find('Father:')!=-1:
                            name_idx=dob_idx-3
                            up_n_idx=True
                    
                
                if up_n_idx==False:
                    name_idx=dob_idx-1
            if flag==2:
                
                name_idx=dob_idx-3
            break
    if flag==0:        
        for i in range(len(text)):
            flag=0
            word_list=text[i].split(' ')
            for word in word_list:
                if word.find('Father')!=-1 or word.find('Father:')!=-1:
                    name_idx=i-1
                    flag=1
                    break
            if flag !=0:
                break
        
            
    return text,dob_idx,name_idx

    
#loading json file
#-------------------------------
#with open('data/json_from_google_vision_aadhaar/a11.json') as json_file: 
with open('./data/json_from_google_vision_aadhaar/a4.json') as json_file:
    data = json.load(json_file) 
    
#fetching raw texts from the JSON
#-----------------------------------
text=data['fullTextAnnotation']['text']
text=text.split('\n')

text=remove_email_address(text)
text,start,end=address_generator(text)

text,dob_idx,name_idx=name_dob_generator(text)

#printing name and DOB
#-----------------------------

print('Fetched Data:',end='\n----------------------\n')

if dob_idx!=-1 :
    print(text[dob_idx])
if name_idx!=-1:
    print('Name: ',text[name_idx])
    

#printing the address
#---------------------------

if(start==-1 and end != -1):
    for i in range(end-5,end+1):
        print(text[i])
        
if(start!=-1 and end != -1):
    for i in range(start,end+1):
        print(text[i])
```
